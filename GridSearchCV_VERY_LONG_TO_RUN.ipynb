```python
# --- Step 3 — Classical Time Series Models (Without pmdarima) ---

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

# Use statsmodels ARIMA instead of pmdarima
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

# ETS (Holt-Winters)
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Prophet
try:
    from prophet import Prophet
except Exception:
    from fbprophet import Prophet

# Diagnostics
from statsmodels.stats.diagnostic import acorr_ljungbox

# ---------------------------
# Metrics & Utilities
# ---------------------------
def rmse(y_true, y_pred):
    return sqrt(mean_squared_error(y_true, y_pred))

def mape(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100.0

def infer_seasonal_period(y: pd.Series):
    """Heuristic seasonal period inference from frequency."""
    if y.index.freqstr is not None:
        freq = y.index.freqstr.upper()
        if freq in ("D", "B"):
            return 7      # weekly seasonality for daily
        if freq in ("H",):
            return 24     # daily seasonality for hourly
        if "W" in freq:
            return 52     # annual for weekly
        if "M" in freq:
            return 12     # annual for monthly
    return 7              # safe default for daily

def rolling_origin_splits(y: pd.Series, n_splits=5, test_horizon=7, min_train_size=0.5):
    """Yield (train_idx, test_idx) pairs for rolling-origin CV."""
    n = len(y)
    start_train = int(n * min_train_size)
    split_points = np.linspace(start_train, n - test_horizon, n_splits, dtype=int)
    for sp in split_points:
        train_idx = np.arange(0, sp)
        test_idx = np.arange(sp, sp + test_horizon)
        if test_idx[-1] < n:
            yield train_idx, test_idx

def ljung_box_pvalue(residuals, lags=12):
    try:
        lb = acorr_ljungbox(residuals, lags=[lags], return_df=True)
        return float(lb['lb_pvalue'].iloc[0])
    except Exception:
        return np.nan

def adf_pvalue(series):
    try:
        return adfuller(series, autolag="AIC")[1]
    except Exception:
        return np.nan

def plot_forecasts(y, forecasts: dict, title="Forecast Comparison"):
    plt.figure(figsize=(12, 6))
    y.plot(label="Actual", color='black', linewidth=2)
    step = y.index[1] - y.index[0]
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    for i, (name, f) in enumerate(forecasts.items()):
        idx = pd.date_range(start=y.index[-1] + step, periods=len(f), freq=y.index.freq)
        pd.Series(f, index=idx).plot(label=name, color=colors[i % len(colors)], linewidth=2)
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# ---------------------------
# Manual Auto-ARIMA Alternative
# ---------------------------
def auto_arima_alternative(y_train, seasonal_period=7, max_p=3, max_d=2, max_q=3, max_P=2, max_D=1, max_Q=2):
    """Grid search alternative to pmdarima's auto_arima"""
    best_aic = np.inf
    best_model = None
    best_order = None
    best_seasonal_order = None

    # Check if data needs differencing
    adf_stat, adf_pval = adfuller(y_train)[:2]
    d_max = 1 if adf_pval < 0.05 else max_d

    print(f"Searching ARIMA parameters (seasonal_period={seasonal_period})...")

    # Test different ARIMA orders
    for p in range(0, max_p + 1):
        for d in range(0, d_max + 1):
            for q in range(0, max_q + 1):
                if seasonal_period > 1:
                    for P in range(0, max_P + 1):
                        for D in range(0, max_D + 1):
                            for Q in range(0, max_Q + 1):
                                try:
                                    model = ARIMA(y_train, order=(p,d,q),
                                                seasonal_order=(P,D,Q,seasonal_period))
                                    fitted = model.fit()
                                    if fitted.aic < best_aic:
                                        best_aic = fitted.aic
                                        best_model = fitted
                                        best_order = (p,d,q)
                                        best_seasonal_order = (P,D,Q,seasonal_period)
                                except:
                                    continue
                else:
                    try:
                        model = ARIMA(y_train, order=(p,d,q))
                        fitted = model.fit()
                        if fitted.aic < best_aic:
                            best_aic = fitted.aic
                            best_model = fitted
                            best_order = (p,d,q)
                            best_seasonal_order = None
                    except:
                        continue

    print(f"Best ARIMA order: {best_order}, seasonal: {best_seasonal_order}, AIC: {best_aic:.2f}")
    return best_model, best_order, best_seasonal_order

# ---------------------------
# Model Wrappers
# ---------------------------
def fit_predict_arima(y_train, fh, seasonal_period):
    model, order, seasonal_order = auto_arima_alternative(y_train, seasonal_period)
    if model is None:
        # Fallback to simple ARIMA(1,1,1)
        try:
            model = ARIMA(y_train, order=(1,1,1)).fit()
        except:
            model = ARIMA(y_train, order=(0,1,0)).fit()

    fc = model.forecast(steps=fh)
    return model, np.asarray(fc)

def fit_predict_ets_grid(y_train, fh, seasonal_period):
    trend_opts = [None, 'add', 'mul']
    seasonal_opts = [None, 'add', 'mul'] if seasonal_period and seasonal_period > 1 else [None]
    best = {"cfg": None, "err": np.inf, "model": None, "fc": None}

    for trend in trend_opts:
        for seasonal in seasonal_opts:
            try:
                model = ExponentialSmoothing(
                    y_train,
                    trend=trend,
                    seasonal=seasonal,
                    seasonal_periods=seasonal_period if seasonal else None,
                    initialization_method='estimated'
                ).fit(optimized=True, use_brute=True)
                fc = model.forecast(fh)
                tail = min(14, len(y_train))
                err = mean_absolute_error(y_train[-tail:], model.fittedvalues[-tail:])
                if err < best["err"]:
                    best = {"cfg": (trend, seasonal), "err": err, "model": model, "fc": np.asarray(fc)}
            except Exception as e:
                continue

    if best["model"] is None:
        # Fallback to simple exponential smoothing
        try:
            model = ExponentialSmoothing(y_train, trend=None, seasonal=None).fit()
            fc = model.forecast(fh)
            best = {"model": model, "fc": np.asarray(fc)}
        except:
            # Ultimate fallback - naive forecast
            best = {"model": None, "fc": np.full(fh, y_train.iloc[-1])}

    return best["model"], best["fc"]

def fit_predict_prophet(y_train: pd.Series, fh: int, holidays_df=None):
    df = pd.DataFrame({"ds": y_train.index, "y": y_train.values})
    m = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        seasonality_mode='additive'
    )
    if holidays_df is not None and len(holidays_df) > 0:
        m.holidays = holidays_df
    m.fit(df)
    future = m.make_future_dataframe(periods=fh, freq=y_train.index.freq)
    fc = m.predict(future)
    fc_tail = fc.tail(fh)["yhat"].values
    return m, np.asarray(fc_tail)

# ---------------------------
# Cross-Validation & Evaluation
# ---------------------------
def evaluate_models_cv(y: pd.Series, n_splits=3, test_horizon=7, holidays_df=None):
    seasonal_period = infer_seasonal_period(y)
    rows = []
    models_per_split = []

    print(f"Starting {n_splits}-fold cross-validation with horizon={test_horizon}")

    for i, (tr, te) in enumerate(rolling_origin_splits(y, n_splits=n_splits, test_horizon=test_horizon)):
        print(f"\n--- Split {i+1}/{n_splits} ---")
        y_tr, y_te = y.iloc[tr], y.iloc[te]

        # Fit models
        print("Fitting ARIMA...")
        arima_model, arima_fc = fit_predict_arima(y_tr, fh=len(y_te), seasonal_period=seasonal_period)

        print("Fitting ETS...")
        ets_model, ets_fc = fit_predict_ets_grid(y_tr, fh=len(y_te), seasonal_period=seasonal_period)

        print("Fitting Prophet...")
        prophet_model, prophet_fc = fit_predict_prophet(y_tr, fh=len(y_te), holidays_df=holidays_df)

        preds = {
            "arima": arima_fc,
            "ets": ets_fc,
            "prophet": prophet_fc,
        }

        # Simple average ensemble
        preds["avg_ensemble"] = (arima_fc + ets_fc + prophet_fc) / 3.0

        # Weighted ensemble based on MAE
        maes = {name: mean_absolute_error(y_te, p) for name, p in preds.items()
                if name in ["arima", "ets", "prophet"]}
        inv = np.array([1.0/(maes[k]+1e-9) for k in ["arima", "ets", "prophet"]])
        wts = inv / inv.sum()
        w_ens = wts[0]*preds["arima"] + wts[1]*preds["ets"] + wts[2]*preds["prophet"]
        preds["w_ensemble"] = w_ens

        # Record metrics
        for name, p in preds.items():
            rows.append({
                "split": i+1,
                "model": name,
                "MAE": mean_absolute_error(y_te, p),
                "RMSE": rmse(y_te, p),
                "MAPE": mape(y_te, p)
            })

        models_per_split.append({
            "split": i+1,
            "arima": arima_model,
            "ets": ets_model,
            "prophet": prophet_model
        })

    metrics = pd.DataFrame(rows)
    leaderboard = (metrics.groupby("model")
                            .agg({"MAE":"mean","RMSE":"mean","MAPE":"mean"})
                            .sort_values("MAE"))
    return metrics, leaderboard, models_per_split, seasonal_period

# ---------------------------
# Fit Final on All Data & Forecast
# ---------------------------
def fit_all_and_forecast(y: pd.Series, horizon=14, holidays_df=None):
    seasonal_period = infer_seasonal_period(y)

    print("Fitting final models on full dataset...")
    arima_model, arima_fc = fit_predict_arima(y, fh=horizon, seasonal_period=seasonal_period)
    ets_model, ets_fc = fit_predict_ets_grid(y, fh=horizon, seasonal_period=seasonal_period)
    prophet_model, prophet_fc = fit_predict_prophet(y, fh=horizon, holidays_df=holidays_df)

    preds = {
        "arima": arima_fc,
        "ets": ets_fc,
        "prophet": prophet_fc,
        "avg_ensemble": (arima_fc + ets_fc + prophet_fc) / 3.0,
    }

    # Calculate in-sample MAE for weighting
    tail = min(14, len(y))
    in_mae = {}

    try:
        in_mae["arima"] = mean_absolute_error(y[-tail:], arima_model.fittedvalues[-tail:])
    except:
        in_mae["arima"] = np.median(np.abs(np.diff(y[-tail:])))

    try:
        in_mae["ets"] = mean_absolute_error(y[-tail:], pd.Series(ets_model.fittedvalues).reindex_like(y).iloc[-tail:])
    except:
        in_mae["ets"] = np.median(np.abs(np.diff(y[-tail:])))

    try:
        df = pd.DataFrame({"ds": y.index, "y": y.values})
        pf_fit = prophet_model.predict(df)["yhat"].values
        in_mae["prophet"] = mean_absolute_error(y[-tail:], pf_fit[-tail:])
    except:
        in_mae["prophet"] = np.median(np.abs(np.diff(y[-tail:])))

    inv = np.array([1.0/(in_mae[k]+1e-9) for k in ["arima", "ets", "prophet"]])
    wts = inv / inv.sum()
    preds["w_ensemble"] = wts[0]*preds["arima"] + wts[1]*preds["ets"] + wts[2]*preds["prophet"]

    return preds, {"arima": arima_model, "ets": ets_model, "prophet": prophet_model}, seasonal_period, wts

# ---------------------------
# Diagnostics
# ---------------------------
def quick_diagnostics(y: pd.Series, model_name: str, model_obj):
    print(f"\n--- Diagnostics for {model_name} ---")
    try:
        if model_name == "arima" and model_obj is not None:
            resid = model_obj.resid
        elif model_name == "ets" and model_obj is not None:
            resid = y - pd.Series(model_obj.fittedvalues).reindex_like(y)
        elif model_name == "prophet" and model_obj is not None:
            df = pd.DataFrame({"ds": y.index, "y": y.values})
            fit = model_obj.predict(df)["yhat"].values
            resid = y.values - fit
        else:
            resid = None

        if resid is not None:
            resid_clean = pd.Series(resid).fillna(0)
            print(f"ADF p-value (stationarity, lower is better): {adf_pvalue(resid_clean):.4f}")
            print(f"Ljung-Box p-value (autocorr, higher ~ white noise): {ljung_box_pvalue(resid_clean.dropna(), lags=12):.4f}")

            plt.figure(figsize=(12,4))
            plt.plot(resid_clean)
            plt.title(f"Residuals: {model_name}")
            plt.grid(True, alpha=0.3)
            plt.show()
    except Exception as e:
        print(f"Diagnostic error: {e}")

# ---------------------------
# Main Execution
# ---------------------------
if __name__ == "__main__":
    print("=== Classical Time Series Analysis (Without pmdarima) ===")

    # Load your data
    try:
        df = pd.read_csv('updated_final_merged_data.csv', index_col='date', parse_dates=True)
        y = df['calls'].asfreq('D')
        print(f"Loaded data: {len(y)} observations from {y.index[0]} to {y.index[-1]}")

        # 1) Cross-validation evaluation
        print("\n" + "="*50)
        print("CROSS-VALIDATION EVALUATION")
        print("="*50)

        metrics, leaderboard, models_splits, seasonal_period = evaluate_models_cv(
            y, n_splits=3, test_horizon=7, holidays_df=None
        )

        print(f"\nInferred seasonal period: {seasonal_period}")
        print("\nLeaderboard (average across splits):")
        print(leaderboard.round(4))

        # 2) Final models and forecasting
        print("\n" + "="*50)
        print("FINAL MODELS & FORECASTING")
        print("="*50)

        horizon = 14
        preds, models, seasonal_period, weights = fit_all_and_forecast(
            y, horizon=horizon, holidays_df=None
        )

        print(f"\nInverse-MAE ensemble weights [arima, ets, prophet]: {weights.round(3)}")
        print(f"\nFinal Forecasts (first 5 periods):")
        for name, fc in preds.items():
            print(f"{name:15}: {np.round(fc[:5], 2)}")

        # 3) Visualization
        plot_forecasts(
            y.tail(90),  # Last 90 days
            preds,
            title="Model Forecasts vs Actuals (Last 90 Days)"
        )

        # 4) Diagnostics
        print("\n" + "="*50)
        print("MODEL DIAGNOSTICS")
        print("="*50)

        for nm, obj in models.items():
            quick_diagnostics(y, nm, obj)

        print("\n=== Analysis Complete ===")

    except FileNotFoundError:
        print("Error: 'updated_final_merged_data.csv' not found. Please upload your data file.")
    except Exception as e:
        print(f"Error: {e}")

        # Demo with synthetic data
        print("\nRunning with synthetic demo data...")
        dates = pd.date_range('2023-01-01', periods=365, freq='D')
        trend = np.linspace(100, 200, 365)
        seasonal = 10 * np.sin(2 * np.pi * np.arange(365) / 7)
        noise = np.random.normal(0, 5, 365)
        y = pd.Series(trend + seasonal + noise, index=dates, name='calls')

        # Run analysis with demo data
        metrics, leaderboard, _, _ = evaluate_models_cv(y, n_splits=3, test_horizon=7)
        print("\nDemo Leaderboard:")
        print(leaderboard.round(4))
```
